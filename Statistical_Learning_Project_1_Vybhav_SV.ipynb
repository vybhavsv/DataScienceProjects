```python
z=Laliga_df['Points'].groupby(Laliga_df['BestPosition']).sum()
Teams_Best_Position = pd.DataFrame(z)
print(Teams_Best_Position.dtypes)
Teams_Best_Position
```

```python
import matplotlib.pyplot as plt
Teams_Best_Position.sort_values(['BestPosition','Points']).plot.bar(color='b')
plt.ylabel('Sum Of Points')
plt.xlabel('Best Position')
plt.xticks(rotation=360)
plt.title('Teams Postions Versus Points')
plt.show()
```
